```python
import wave
import os
```

```python

for file_name in os.listdir('.'):
    print(f"File: {file_name}")
    if not file_name.endswith('.wav'):
        continue
    with wave.open(file_name, 'rb') as audio_file:
        print(f"File: {file_name}")
        channels = audio_file.getnchannels()
        sample_width = audio_file.getsampwidth()
        frame_rate = audio_file.getframerate()
        bit_rate_bps = channels * sample_width * 8 * frame_rate
        # Convert to kbps
        bit_rate_kbps = bit_rate_bps / 1000
        print(f"Channels: {channels}")
        print(f"Sample Width: {sample_width}")
        print(f"Frame Rate (Sample Rate): {audio_file.getframerate()}")
        print(f"Bit Rate: {bit_rate_kbps} kbps")
        print(f"Number of Frames: {audio_file.getnframes()}")
        print(f"Compression Type: {audio_file.getcomptype()} \n")
        #if bit_rate_kbps <2000:
         #   os.remove(file_name)
```
